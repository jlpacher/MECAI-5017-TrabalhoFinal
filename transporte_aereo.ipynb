Com o intuito de ampliar o conhecimento da sociedade brasileira e de subsidiar a realização de pesquisas, estudos e análises mais abrangentes sobre o setor, a ANAC tem disponibilizado, na seção "Dados e Estatísticas" do seu portal na internet, relatórios, estudos e informações sobre as condições de mercado. <br>

Nesse sentido, a Agência agrega, para a livre consulta de qualquer interessado, a série histórica dos dados estatísticos do transporte aéreo do Brasil, com elevado grau de detalhamento. <br>

Estão contempladas informações sobre a quantidade de passageiro, carga e mala postal transportados, distância voada, combustível consumido, entre outras, por etapa de voo e por empresa aérea. <br>

Essas informações podem ser utilizadas para apurar importantes indicadores do setor, como demanda (RPK, RTK), oferta (ASK, ATK), participação de mercado, taxa de ocupação das aeronaves (Load Factor), entre outros. <br>

Os dados estatísticos do transporte aéreo do Brasil encontram-se regulamentados pela Resolução ANAC no 191/2011 e pelas Portarias ANAC no 1.189 e 1.190/SRE/2011. <br>

De acordo com a mencionada regulamentação, os dados são mensalmente fornecidos à ANAC, até o dia 10 do mês subsequente ao de referência, pelas empresas brasileiras e estrangeiras que exploram os serviços de transporte aéreo público regular e não regular no Brasil. <br>

Na busca pela melhoria continuada da qualidade da informação e com o propósito de alcançar o maior nível de consistência possível, os dados são submetidos a críticas, validações e procedimentos de auditoria pela Agência. Assim, os dados estão sujeitos a revisões, correções e alterações, podendo apresentar diferenças em relação àqueles divulgados anteriormente ou mesmo discrepâncias e observações, conhecidas como outliers, que devem ser consideradas em sua análise.​ <br>

Metadados <br>
Dados Estatísticos do Transporte Aéreo <br>​
​
Etapa Básica (flight stage) <br>

As etapas básicas são aquelas realizadas pela aeronave desde a sua decolagem até o próximo pouso, independente de onde tenha sido realizado o embarque ou o desembarque do objeto de transporte. Os dados estatísticos das etapas básicas representam o status da aeronave em cada etapa do voo, apresentando a movimentação de cargas e passageiros entre os aeródromos de origem e destino da aeronave. É a operação de uma aeronave entre uma decolagem e o próximo pouso, ou seja, é a ligação direta entre dois aeródromos. Os aeroportos correspondem aos campos AEROPORTO DE ORIGEM (SIGLA);AEROPORTO DE ORIGEM (NOME);AEROPORTO DE ORIGEM (UF);AEROPORTO DE ORIGEM (REGIÃO);AEROPORTO DE ORIGEM (PAÍS);AEROPORTO DE ORIGEM (CONTINENTE);AEROPORTO DE DESTINO (SIGLA);AEROPORTO DE DESTINO (NOME);AEROPORTO DE DESTINO (UF);AEROPORTO DE DESTINO (REGIÃO);AEROPORTO DE DESTINO (PAÍS);AEROPORTO DE DESTINO (CONTINENTE) do arquivo. <br>

Campo	Descrição <br>
ATK (Available tonne kilometer)	Refere-se ao volume de Tonelada Quilômetro Oferecida, ou seja, a soma do produto entre o payload, que é a capacidade total de peso disponível na aeronave, expressa em quilogramas, disponível para efetuar o transporte de passageiros, carga e correio, e a distância das etapas, dividido por 1.000. <br>
RTK (Revenue tonne kilometer):	Refere-se ao volume de Toneladas Quilômetros Transportadas, ou seja, a soma do produto entre os quilogramas carregados pagos, onde cada passageiro possui o peso estimado de 75 Kg, e a distância das etapas, dividido por 1.000. <br>
Combustível	Refere-se à quantidade, em litros, de combustível consumida pela aeronave na execução da referida etapa. Informação disponível apenas para empresas brasileiras. <br>
Distância	Refere-se à distância, expressa em quilômetros, entre os aeródromos de origem e destino da etapa, considerando a curvatura do planeta Terra. <br>
Horas Voadas	Refere-se ao número de horas de voo entre os aeródromos de origem e destino da etapa. <br>
Decolagens	Refere-se ao número de decolagens que ocorreram entre os aeródromos de origem e destino da etapa. <br>
Carga Paga Km	Refere-se ao volume de Carga Paga (kg) em cada quilômetro, ou seja, a soma do produto entre a quantia (kg) de carga paga e a distâncias das etapas. <br>
Carga Grátis Km	Refere-se ao volume de Carga Grátis (kg) em cada quilômetro, ou seja, a soma do produto entre a quantia (kg) de carga grátis e a distâncias das etapas. <br>
Correio Km	Refere-se ao volume de Correio (kg) em cada quilômetro, ou seja, a soma do produto entre a quantia (kg) de correio e a distâncias das etapas. <br>
Assentos	É o número de assentos disponíveis em cada etapa de voo de acordo com a configuração da aeronave na execução da etapa; e <br>
Payload (Kg) (Payload capacity)	é a capacidade total de peso na aeronave, expressa em quilogramas, disponível para efetuar o transporte de passageiros, carga e correio. <br>
Etapa Combinada (On flight origin and destination - OFOD) <br>

As etapas combinadas identificam os pares de aeródromos de origem, onde houve o embarque do objeto de transporte, e destino, onde houve o desembarque do objeto de transporte, independente da existência de aeródromos intermediários, atendidos por determinado voo. É a etapa de voo vista com foco no objeto de transporte (pessoas e/ou cargas), com base no embarque e desembarque nos aeródromos relacionados. Os dados estatísticos da etapa combinada informam a origem e destino no voo, dos passageiros e cargas transportadas, independente das suas escalas. Os aeroportos correspondem aos campos AEROPORTO DE ORIGEM (SIGLA);AEROPORTO DE ORIGEM (NOME);AEROPORTO DE ORIGEM (UF);AEROPORTO DE ORIGEM (REGIÃO);AEROPORTO DE ORIGEM (PAÍS);AEROPORTO DE ORIGEM (CONTINENTE);AEROPORTO DE DESTINO (SIGLA);AEROPORTO DE DESTINO (NOME);AEROPORTO DE DESTINO (UF);AEROPORTO DE DESTINO (REGIÃO);AEROPORTO DE DESTINO (PAÍS);AEROPORTO DE DESTINO (CONTINENTE) do arquivo. <br>

Campo	Descrição <br>
Empresa Aérea	Empresa Aérea responsável por operar as etapas. Corresponde aos campos EMPRESA (SIGLA);EMPRESA (NOME);EMPRESA (NACIONALIDADE) do arquivo.
Ano e Mês	Refere-se à data prevista para início da primeira etapa de cada voo. <br>
Natureza do Voo	<br>
Refere-se à natureza das etapas, e possui o valor “Doméstico" caso as etapas tenham o pouso e a decolagem realizadas no Brasil e sejam operadas por Empresas brasileiras ou possuem o valor “Internacional" caso contrário. <br>
Tipo de Voo	<br>
Faz referência ao tipo de operação das etapas: <br>

Improdutivas (Non-revenue flights): etapas que não geraram receita à empresa aérea (como realização de treinamentos, voo para manutenção de aeronaves); <br>

Regulares (Scheduled revenue flights): etapas remuneradas que são realizadas sob uma numeração de Horário de Transporte (HOTRAN). Recebem esse nome, pois possuem a característica de serem realizadas regularmente; e <br>

Não Regulares (Non-scheduled revenue flights): etapas remuneradas que não são realizadas sob uma numeração de Horário de Transporte (HOTRAN). Recebem esse nome, pois possuem a característica de serem realizadas de forma não continuada. Aqui estão os voos Charters, Fretamentos. <br>

Passageiros Pagos <br>
Refere-se aos passageiros que ocupam assentos comercializados ao público e que geram receita, com a compra de assentos, para a empresa de transporte aéreo. Incluem-se nesta definição as pessoas que viajam em virtude de ofertas promocionais, as que se valem dos programas de fidelização de clientes, as que se valem dos descontos concedidos pelas empresas, as que viajam com tarifas preferenciais, as pessoas que compram passagem no balcão ou através do site de empresa de transporte aéreo e as pessoas que compram passagem em agências de viagem. <br>

Passageiros Grátis <br>
Refere-se aos passageiros que ocupam assentos comercializados ao público mas que não geram receita, com a compra de assentos, para a empresa de transporte aéreo. Incluem-se nesta definição as pessoas que viajam gratuitamente, as que se valem dos descontos de funcionários das empresas aéreas e seus agentes, os funcionários de empresas aéreas que viajam a negócios pela própria empresa e os tripulantes ou quem estiver ocupando assento destinado a estes. <br>

Carga Paga <br>
Refere-se à quantidade total, expressa em quilogramas, de todos os bens que tenham sido transportados na aeronave, exceto correio e bagagem, e tenham gerado receitas direta ou indireta para a empresa aérea. <br>

Carga Grátis <br>
Refere-se à quantidade total, expressa em quilogramas, de todos os bens que tenham sido transportados na aeronave, exceto correio e bagagem, e não tenha gerado receitas diretas ou indiretas para a empresa aérea. <br>

Correio	<br>
Refere-se à quantidade de objetos transportados de rede postal em cada trecho de voo realizado, expresso em quilogramas. <br>

Bagagem	<br>
Refere-se à quantidade total de bagagem despachada, expressa em quilogramas. <br>

​Obs.: Os Dados Estatísticos possuem grande semelhança em relação ao programa estatístico da International Civil Aviation Organization (ICAO), em especial no significado das variáveis. Assim, para facilitar a comparação, os nomes de algumas variáveis são apresentados também em inglês. <br>

Apesar de estarem dispostas como informações de etapas básicas, as variáveis RPK, RTK, Carga Paga Km, Carga Grátis Km e Correio Km das empresas estrangeiras são computadas por meio das informações congêneres advindas das etapas combinadas. Assim, ao se desagregar essas variáveis de demanda das empresas estrangeiras por aeroporto, ou por rota, e as comparar com as informações de oferta (ASK e ATK, originadas das etapas básicas), é possível que os valores das taxas de aproveitamento estejam subdimensionados ou superdimensionados, e, em alguns casos, estarem superiores a 100%. Por exemplo, o Load Factor (RPK/ASK) de empresas estrangeiras, desagregado por aeroporto, pode ser superior a 100%, devido à formatação dos dados. <br>

Passenger Kilometre/ Tonne Kilometre – Metrics used in the aviation industry to measure performance: <br>
Available Seat Kilometres (ASK) – a measure of an airline’s carrying capacity to generate revenue, taken from multiplying the available seats on any given aircraft by the number of kilometres flown on a given flight. <br>
Revenue Passenger Kilometres (RPK) – a way of calculating the number of kilometres travelled by paying customers, by multiplying the number of paying passengers by the distance travelled in kilometers. <br>
Tonne Kilometres (TKМ) – a way of measuring an airline’s cargo and passenger carriage capacity – the movement of one tonne of cargo (or passengers, calculated at 90 Kg per passenger, and commercial cargo) one kilometre, measured in tonne kilometres. <br>
Passenger Load Factor (PLF) – a measure for an airline’s capacity, defined as the ratio between RPK and ASK. <br>

```python
# Calcula a quantidade de nulos em cada coluna
null_counts = df.isnull().sum().reset_index()
null_counts.columns = ['coluna', 'quantidade_nulos']

# Calcula os percentuais de valores nulos em cada coluna
total_rows = len(df)
null_counts['percentual_nulos'] = null_counts['quantidade_nulos'] / total_rows * 100

# Configura o tamanho do gráfico
plt.figure(figsize=(10, 6))

# Define uma paleta de cores personalizada
custom_palette = sns.color_palette("coolwarm", as_cmap=True)

# Normaliza os percentuais de valores nulos para a paleta de cores
norm = plt.Normalize(null_counts['percentual_nulos'].min(), null_counts['percentual_nulos'].max())
sm = plt.cm.ScalarMappable(cmap=custom_palette, norm=norm)
sm.set_array([])

# Converte os valores normalizados em uma lista de cores
colors = list(custom_palette(norm(null_counts['percentual_nulos'])))

# Gera o gráfico de barras usando seaborn com a paleta de cores baseada nos percentuais de nulos
ax = sns.barplot(x='coluna', y='quantidade_nulos', data=null_counts, palette=colors, hue='coluna', legend=False)

# Adiciona uma barra de cores para indicar os percentuais
cbar = plt.colorbar(sm, ax=ax, label='Percentual de Valores Nulos')

# Adiciona título e rótulos aos eixos
plt.title('Quantidade de Valores Nulos por Coluna', fontweight='bold')
plt.xlabel('')
plt.ylabel('Quantidade de Valores Nulos')

# Gira os rótulos das colunas para melhor visualização
plt.xticks(rotation=90)

# Exibe o gráfico
plt.show()
```

```python
result = df.groupby('ANO').apply(lambda x: pd.Series({
    'Exemplos': x.shape[0],
    'Total': x.size,
    'Nulos': x.isnull().sum().sum(),
    'Percentual de Nulos (%)': (x.isnull().sum().sum() / x.size) * 100
})).reset_index()

result = result.loc[result['ANO'] < 2024]  # Filtrar para os anos desejados

# Cores fornecidas navy black
colors = ['steelblue', 'steelblue', 'steelblue', 'navy', 'navy', 'darkred',
          'darkred', 'darkred', 'yellow', 'lightsalmon', 'lightsalmon', 'lightsalmon', 'lightsalmon',
          'lightsalmon', 'lightsalmon', 'lightsalmon', 'lightsalmon',
          'orange', 'red', 'red', 'red', 'red', 'orange', 'lightsalmon']

# Inicializar o layout do gráfico
fig, ax = plt.subplots(figsize=(9, 6))

# Adicionar fundo cinza no eixo e na figura
ax.set_facecolor('whitesmoke')
fig.set_facecolor('whitesmoke')

# Inverter a ordem dos anos no eixo Y
result_sorted = result.sort_values('ANO', ascending=False)

# Criar o gráfico de barras horizontais
bars = ax.barh(result_sorted['ANO'], result_sorted['Percentual de Nulos (%)'], color=colors, zorder=2)

# Adicionar uma grade vertical leve para facilitar a leitura
ax.grid(linestyle='-', alpha=0.5, axis='x')

# Título do gráfico
title = 'Histograma \nDistribuição do Percentual de Nulos por Ano'
fig.text(0.02, 1.02, title, fontsize=25, fontweight='bold', ha='left', family='dejavu sans')

# Subtítulo do gráfico
# subtitle = 'Histograma \nPercentual de valores nulos em registros ao longo dos anos'
# fig.text(0.02, 0.97, subtitle, fontsize=13, color='dimgrey', ha='left', family='dejavu sans')

# Detalhes e Créditos (ajustar conforme necessário)
# text = 'Fonte: Dados fictícios para exemplo'
# fig.text(0.02, 0.02, text, fontsize=10, color='dimgrey', ha='left', family='dejavu sans')

# Remover as bordas superiores e direitas do gráfico para um estilo mais limpo
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Ajustar os parâmetros dos eixos
ax.tick_params(axis='y', labelsize=12)
ax.xaxis.tick_top()

# Remover os rótulos padrão do eixo X
ax.set_xticklabels([])

# Ajustar a posição do eixo X
ax.set_xlim(0, result_sorted['Percentual de Nulos (%)'].max() + 1)

# Adicionar os valores à direita de cada barra
for i in range(len(result_sorted)):
    ax.text(result_sorted['Percentual de Nulos (%)'].iloc[i] + 0.1,  # posição X ligeiramente à direita da barra
            result_sorted['ANO'].iloc[i],  # posição Y alinhada ao centro da barra
            f"{result_sorted['Percentual de Nulos (%)'].iloc[i]:.1f}%",  # texto com o valor percentual
            va='center',  # alinhamento vertical
            fontsize=9,  # tamanho da font
            color='black')  # cor do texto

# Exibir o gráfico final
plt.show()
```

```python
# Função para criar subplots de pizza
def gerar_subplots_pizza(variavel, titulo, ordem_periodos):
  """
  Gera subplots de pizza para a variável especificada, agrupado por região e faixa de anos.

  Args:
    data: Dataframe com os dados.
    variavel: Nome da variável a ser plotada.
    titulo: Título do conjunto de subplots.
    ordem_periodos: Lista com a ordem dos períodos a serem exibidos.

  """
  fig, axes = plt.subplots(1, 3, figsize=(9, 3))

  for i, periodo in enumerate(ordem_periodos):
    # Filtrar dados por faixa de anos
    anos_inicio, anos_fim = periodo.split(' a ')
    data_filtrada = result[(result['ANO'] >= int(anos_inicio)) & (result['ANO'] <= int(anos_fim))]

    # Agrupar por região e somar valores
    data_agrupada = data_filtrada.groupby('AEROPORTO_DE_ORIGEM_REGIAO')[variavel].sum().reset_index()

    # Criar gráfico de pizza no subplot
    axes[i].pie(data_agrupada[variavel], labels=data_agrupada['AEROPORTO_DE_ORIGEM_REGIAO'], autopct=f"%1.1f%%", textprops={'fontsize': 8})
    axes[i].set_title(f"{periodo}", fontweight='bold')

  fig.suptitle(titulo, fontweight='bold')
  plt.tight_layout()
  plt.show()
  print()

# Gerando subplots para cada variável

# 1) Número de voos
gerar_subplots_pizza("NUMERO_DE_VOOS", "Número de Voos por Região de Origem do Aeroporto", ['2000 a 2009', '2010 a 2019', '2020 a 2023'])

# 2) Passageiros pagos
gerar_subplots_pizza("PASSAGEIROS_PAGOS_SUM", "Passageiros Pagos por Região de Origem do Aeroporto", ['2000 a 2009', '2010 a 2019', '2020 a 2023'])

# 3) Carga paga em kg
gerar_subplots_pizza("CARGA_PAGA_KG_SUM", "Carga Paga por Região de Origem do Aeroporto", ['2000 a 2009', '2010 a 2019', '2020 a 2023'])
```

```python
# Passenger Load Factor (PLF) – a measure for an airline’s capacity, defined as the ratio between RPK and ASK.

data = df.copy()
data['PLF'] = data['RPK'] / data['ASK']
data['PLF'] = pd.to_numeric(data['PLF'], errors='coerce')
data['PLF'].replace([np.inf, -np.inf], np.nan, inplace=True)
data['PLF'].dropna(inplace=True)
data = data.loc[(data['PLF']>=0.0) & (data['PLF']<=1.0)]

# Sort data by 'PLF' column in ascending order
data_sorted = data.sort_values(by='PLF')

# Group data by 'PLF' and calculate frequencies
grouped_data = data_sorted.groupby('PLF')['PLF'].size()

# Calculate weighted cumulative probabilities
cdf_values = np.cumsum(grouped_data.values) / grouped_data.sum()

# Extract sorted 'PLF' values for x-axis
plf_values = grouped_data.index.to_numpy()

plt.figure(figsize=(10, 6))  # Adjust figure size as needed

# Plot CDF curve
plt.plot(plf_values, cdf_values, linestyle='-', linewidth=3, color='skyblue')

# Set labels and title
plt.xlabel('Fator de Carga')
plt.ylabel('Frequência Acumulada')
plt.title("Função de Distribuição Acumulada \n Fator de Carga", fontweight='bold')

# Show the plot
plt.grid(True)
plt.show()
```

```python
data = df.copy()
data['PLF'] = data['RPK'] / data['ASK']
data['PLF'] = pd.to_numeric(data['PLF'], errors='coerce')
data['PLF'].replace([np.inf, -np.inf], np.nan, inplace=True)
data = data.loc[(data['PLF']>=0.0) & (data['PLF']<=1.0)]
data_filtered = data[data['PLF'].notna()]  # Remove NaN values

# Define time period filters and colors
period_filters = [
    (data_filtered['ANO'] >= 2000) & (data_filtered['ANO'] <= 2007),
    (data_filtered['ANO'] >= 2008) & (data_filtered['ANO'] <= 2015),
    (data_filtered['ANO'] >= 2016) & (data_filtered['ANO'] <= 2023),
]
colors = ['skyblue', 'lightgreen', 'lightcoral']

# Create a single figure for all plots
plt.figure(figsize=(12, 8))  # Adjust figure size as needed

# Iterate over time periods and plot CDFs
for i, (period_filter, color) in enumerate(zip(period_filters, colors)):
    data_period = data_filtered[period_filter]

    # Group data by 'PLF' and calculate frequencies
    grouped_data = data_period.groupby('PLF')['PLF'].size()

    # Calculate weighted cumulative probabilities (CDF) with a tolerance for rounding errors
    tolerance = 1e-6  # Adjust tolerance as needed (e.g., smaller for higher precision)
    cdf_values = np.cumsum(grouped_data.values) / grouped_data.sum()
    cdf_values = np.clip(cdf_values, 0.0, 1.0 - tolerance)  # Clip to ensure final value is close to 1

    # Extract sorted 'PLF' values for x-axis
    plf_values = grouped_data.index.to_numpy()

    # Plot CDF curve for the current period
    plt.plot(plf_values, cdf_values, linestyle='-', linewidth=3, label=f"{2000 + 8 * i} - {2007 + 8 * i}", color=color)

# Set labels and title
plt.xlabel('Fator de Carga')
plt.ylabel('Frequência Acumulada')
plt.title("Função de Distribuição Acumulada \n Fator de Carga", fontweight='bold')

# Add grid lines for better visualization
plt.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
